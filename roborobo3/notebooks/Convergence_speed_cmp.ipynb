```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from os.path import join
sns.set({'figure.figsize': (10, 7)})
```

```python
tot = pd.DataFrame()
for es in ['cmaes', 'fitprop']:
    for it in range(8):
        p = '../logs/conv_speed/{}_rep_{}/fitnesslog.txt'.format(es, it)
        fitlog = pd.read_csv(p, delimiter="\t")
        fitlog['es'] = es
        fitlog['run'] = it
        tot = tot.append(fitlog, ignore_index=True)
```

```python
sns.tsplot(data=tot.query("gen < 10000"), time='gen', condition='es', value='medfit', unit='run', err_style="unit_traces", estimator=np.median)
```
