```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
sns.set()
```

```python
# Methode 2

coef = []
nb = 1000

minquant = scipy.stats.norm.cdf(0.5, loc=1, scale=0.3)
maxquant = scipy.stats.norm.cdf(1.5, loc=1, scale=0.3)
stepquant = (maxquant - minquant) / (nb - 1)
curquant = minquant
for i in range(nb):
    coef.append(scipy.stats.norm.ppf(curquant, loc=1, scale=0.3))
    #coef.append(1/scipy.stats.norm.ppf(curquant, loc=1, scale=0.3))
    curquant += stepquant

fig, ax = plt.subplots(1, 1)
sns.distplot(coef, ax=ax)
plt.show(fig)

print(6.5 / np.mean(coef))
print(6.5 * 1.5)

fig, ax = plt.subplots(1, 1)
sns.swarmplot(coef, ax=ax)
ax.set_xlim(0, 2.1)
plt.show(fig)
```

```python
# methode 1

coef = []
nb=1000
minquant = scipy.stats.norm.cdf(0.5, loc=1, scale=0.3)
maxquant = scipy.stats.norm.cdf(1, loc=1, scale=0.3)
stepquant = (maxquant - minquant) / (nb/2 - 1)
curquant = minquant
for i in range(int(nb/2)):
    coef.append(scipy.stats.norm.ppf(curquant, loc=1, scale=0.3))
    coef.append(1/scipy.stats.norm.ppf(curquant, loc=1, scale=0.3))
    curquant += stepquant

sns.distplot(coef)

print(np.mean(coef))
print(6.5 / np.mean(coef))
```
