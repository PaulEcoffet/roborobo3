```python
import sympy as sp
from sympy import symbols
from sympy.plotting import plot, plot3d
sp.init_printing()
```

```python
x, y = symbols("x y")
```

```python
P = sp.sqrt(2) * sp.sqrt(x + y) - x
P
```

```python
plot(P.subs(y, 0.1), (x, 0, 2))
```

```python
plot3d(P, (x, 0, 2), (y, 0, 2))
```

```python
plot(P.subs(y, x), (x, 0, 2))
```

```python
P.diff(x)
```

```python
sp.solve(P.diff(x))
```

```python
ddp = -10 + y
ddp
```

```python
out = sp.integrate(sp.integrate(ddp, x), x) + 10 
out.subs(x, x - 10)
```

```python
plot(out.subs(y, 2))
```

```python
P = sp.exp(-((0.5) - (x + y))**2/0.05 - x)
plot(P.subs(y, 0.1), (x, 0,2))
```
