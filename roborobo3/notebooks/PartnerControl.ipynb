```python
data = pd.read_csv(folder+"/analysis_log.txt", delimiter="\t")
```

```python
last_gen = data['generation'].max()
last_gen
```

```python
dat = data.query('onOpp == 1 and generation == {}'.format(last_gen)).groupby(("rep", "coop"), as_index=False).mean()
dat
```

```python
def payoff_gauss(own, other):
    return np.exp(-np.power(own + other - 0.5, 2)/0.05);

def payoff_true(own, other):
    coeff = np.sqrt(2)
    return coeff * np.power(own + other, 0.5) - own

def payoff_new(own, other):
    totalInvest = own + other
    invest = own
    a = 3
    B = 2
    q = 1
    n = B
    c= 0.2
    p = B * np.power(totalInvest, a) / (np.power(q, a) + np.power(totalInvest, a))
    share = p / n
    g = np.power(share, a) / (1 + np.power(share, a))
    return g - c * invest
    
    
payoff = payoff_new
dat['reward'] = payoff(dat['ownCoop'], dat['coop'])
dat['totreward'] = payoff(dat['ownCoop'], dat['coop']) * 500
```

```python
dat.query('rank == 1').sum() / 10 * 16
```

```python
sns.violinplot(data=dat, x='coop', y='reward', cut=0)
```

```python
sns.tsplot(data=dat, time='coop', value='reward', unit="ind", err_style="unit_traces", estimator=np.nanmean)
plt.plot(np.linspace(0, 1, 1000), payoff(np.max((0.5-np.linspace(0, 1, 1000), np.zeros(1000)), 0), np.linspace(0, 1, 1000)), c='red', label='Récompense obtenue avec investissement optimal')
plt.ylim((-2, 2))
plt.legend()
plt.ylabel('Récompense obtenue')
plt.xlabel('Investissement de l\'autre')
```

```python
sns.tsplot(data=dat, time='coop', value='ownCoop', unit="rep", err_style="unit_traces", estimator=np.nanmean)
own = np.linspace(0, 2, 500)
other = np.outer(np.linspace(0, 2, 500),(np.ones(own.shape[0])))
best = np.argmax(payoff(own, other), axis=1)
plt.plot(own, own[best].T, c='red', label='propre investissement maximisant la récompense propre')
plt.ylim((-0.1, 2))
plt.legend()
plt.ylabel('Propre investissement')
plt.xlabel("Investissement de l'autre")
plt.title('Réponse moyenne du meilleur individu après {} generations ({} évaluations)'.format(data['generation'].max(), data['generation'].max() *50))
```

```python
pal = sns.palettes.SEABORN_PALETTES['colorblind']
for i, y in enumerate((0.1, 0.3, 0.4)):
    x = np.linspace(0, 2, 1000)
    pay = np.exp(5 * payoff(x, y))
    plt.plot(x, pay, color=pal[i], label='$x_o$ = {}'.format(y))
    pay_max = np.max(pay)
    bound = pay_max * 0.95
    i_ok = np.where(pay > bound)[0]
    xmin = x[i_ok[0]]
    xmax = x[i_ok[-1]]
    plt.axvspan(xmin, xmax, alpha=0.3, color=pal[i], label="$x$ pour $P(x) > 95\% P_{max}$")
plt.legend()
plt.title('Payoff en fonction de l\'investissement propre ($P(x_s, x_o)$)')
plt.xlabel('Propre investissement $x_s$')
plt.ylabel('Gain propre ($P(x_s, x_o)$)')
```
