```python
data = pd.read_csv(folder+"/analysis_log.txt", delimiter="\t")
```

```python
dat = data.query('onOpp == 1').groupby(("rep", "coop"), as_index=False).mean()
dat
```

```python
def payoff_gauss(own, other):
    return np.exp(-np.power(own + other - 0.5, 2)/0.05);

def payoff_true(own, other):
    coeff = np.sqrt(2)
    return coeff * np.power(own + other, 0.5) - own

def payoff_new(own, other):
    totalInvest = own + other
    invest = own
    n = 2
    b = 1
    a = 5
    return (a * totalInvest + b * (totalInvest - invest)) / n - 0.5 * invest * invest
    
    
payoff = payoff_new
dat['reward'] = payoff(dat['ownCoop'], dat['coop'])
dat['totreward'] = payoff(dat['ownCoop'], dat['coop']) * 500
```

```python
sns.violinplot(data=dat, x='coop', y='reward', cut=0)
```

```python
sns.tsplot(data=dat, time='coop', value='ownCoop', unit="rep", err_style="unit_traces", estimator=np.nanmean)
own = np.linspace(0, 10, 500)
other = np.outer(np.linspace(0, 10, 500),(np.ones(own.shape[0])))
best = np.argmax(payoff(own, other), axis=1)
plt.plot(own, own[best].T, c='red', label='propre investissement maximisant la récompense propre')
plt.legend()
plt.ylim((0,10))
plt.ylabel('Propre investissement')
plt.xlabel("Investissement de l'autre")
plt.title('Réponse')
```

```python
pal = sns.palettes.SEABORN_PALETTES['colorblind']
for i, y in enumerate(np.linspace(0, 10, 4)):
    x = np.linspace(0, 10, 1000)
    pay = payoff(x, y)
    plt.plot(x, pay, color=pal[i%len(pal)], label='$x_o$ = {}'.format(y))
    pay_max = np.max(pay)
    bound = pay_max * 0.95
    i_ok = np.where(pay > bound)[0]
    xmin = x[i_ok[0]]
    xmax = x[i_ok[-1]]
    plt.axvspan(xmin, xmax, alpha=0.3, color=pal[i], label="$x$ pour $P(x) > 95\% P_{max}$")
plt.legend()
plt.title('Payoff en fonction de l\'investissement propre ($P(x_s, x_o)$)')
plt.xlabel('Propre investissement $x_s$')
plt.ylabel('Gain propre ($P(x_s, x_o)$)')
```
